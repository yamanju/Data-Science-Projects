## CODING TASK #1. PLOT PIE CHART USING MATPLOTLIB

- **The plot method on Pandas Series and DataFrames is just a simple wrapper around plt.plot():**
- **Define a Pandas Dataframe with all crypto allocation in a given portfolio**
- **Note that total sum = 100%** 
- **Use matplotlib to plot a pie chart** 

```python
# The plot method on Pandas Series and DataFrames is just a simple wrapper around plt.plot():
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
# Define a Pandas Dataframe with all crypto allocation in a given portfolio
# Note that total sum = 100% 
crypto_df = pd.DataFrame(data = {'allocation %':[20, 55, 5, 17, 3]}, index = ['BTC', 'ETH', 'LTC', 'XRP', 'ADA'])
crypto_df
```

```python
# Use matplotlib to plot a pie chart 
crypto_df.plot.pie(y = 'allocation %', figsize = (8, 8))
plt.title('CRYPTO PORTFOLIO PIE CHART');
```

```python
daily_return_df = pd.read_csv('crypto_daily_returns.csv')
mu = round(daily_return_df['BTC'].mean(), 2)  # mean of distribution
sigma = round(daily_return_df['BTC'].std(), 2)  # standard deviation of distribution

num_bins = 40

plt.hist(daily_return_df['BTC'], bins=num_bins, alpha=0.5, color='blue', edgecolor='black')

plt.xlabel('Daily Returns')
plt.ylabel('Frequency')
plt.title('Histogram of BTC Daily Returns')

plt.axvline(x=mu, color='red', linestyle='dashed', linewidth=1, label=f'BTC Mean: {mu}')
plt.axvline(x=sigma, color='black', linestyle='dashed', linewidth=1, label=f'BTC Std: {sigma}')

plt.legend()

plt.show()
#since we have outliers in the histogram std value is bigger than mean value
```

## CODING TASK #5. PLOT SCATTERPLOT & COUNTPLOT USING SEABORN
- **Seaborn is a visualization library that sits on top of matplotlib**

- **Check out the head of the dataframe**

- **Let's print out countplot to know how many samples belong to class #0 and #1**

```python
# import libraries 
import pandas as pd # Import Pandas for data manipulation using dataframes
import numpy as np # Import Numpy for data statistical analysis 
import matplotlib.pyplot as plt # Import matplotlib for data visualisation
import seaborn as sns # Statistical data visualization
```

```python
# Import Cancer data 
cancer_df = pd.read_csv('cancer.csv')
```

```python
cancer_df.head()
```

```python
cancer_df.tail()
```

```python
mean_area = cancer_df['mean area']
mean_smoothness = cancer_df['mean smoothness']

plt.scatter(mean_area, mean_smoothness, alpha=0.5)

plt.xlabel('Mean Area')
plt.ylabel('Mean Smoothness')
plt.title('Scatter Plot: Mean Area vs. Mean Smoothness')

plt.show()
```

```python
#%matplotlib notebook
plt.figure(figsize = (10, 5))

sns.scatterplot(x='mean area', y='mean smoothness',hue='target',palette='plasma', data=cancer_df,s=200)
plt.title('Mean Area vs. Mean Smoothness')
plt.show()
#we do not have enough data to conclude smoothness feature effect the cancer
#on the contrary smaller mean area effect the cancer
```

```python
correlation_matrix = cancer_df.select_dtypes("number").corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, cmap="coolwarm", fmt=".2f");
```

```python
correlation_with_target = correlation_matrix["target"].drop("target")
correlation_with_target = correlation_with_target.sort_values(ascending=False)

plt.figure(figsize=(10, 7))

sns.barplot(x=correlation_with_target.values, y=correlation_with_target.index, palette="coolwarm")

plt.title("Target variable versus other features")

plt.show()
```

```python
ax = sns.countplot(data=cancer_df, x='target')

plt.xlabel('Class')
plt.ylabel('Count')

for p in ax.patches:
    ax.annotate(int(p.get_height()), (p.get_x()+.3, p.get_height()+2));
    
plt.show()
#most of the cases, target is malignant
```

```python
mean_radius = cancer_df['mean radius']
mean_area = cancer_df['mean area']

plt.scatter(mean_radius, mean_area, alpha=0.5)

plt.xlabel('Mean Radius')
plt.ylabel('Mean Area')
plt.title('Scatter Plot: Mean Radius vs. Mean Area')

plt.show()
#heat map also conclude that mean radius and mean area strongly correlated
```

## CODING TASK #6. PLOT PAIRPLOT, DISPLOT, AND HEATMAPS/CORRELATIONS USING SEABORN

- **Plot the pairplot**
- **Strong correlation between the mean radius and mean perimeter, mean area and mean primeter**
- **plot the distplot** 
- **Displot combines matplotlib histogram function with kdeplot() (Kernel density estimate)**
- **KDE is used to plot the Probability Density of a continuous variable.** 

```python
sns.pairplot(data=cancer_df[['mean radius',"mean perimeter","mean area","target",'worst radius',"worst perimeter"]], hue = "target")

plt.show()
```

```python
correlation_matrix = cancer_df.select_dtypes("number").corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, cmap="coolwarm", fmt=".2f");
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(cancer_df[['mean radius',"mean perimeter","mean area"]].corr(), cmap="coolwarm", fmt=".2f",annot = True);
#Strong correlation between the mean radius and mean perimeter, mean area and mean primeter
#worst radius and worst perimetre also strongly correlated
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(cancer_df[['worst radius',"worst perimeter"]].corr(), cmap="coolwarm", fmt=".2f",annot = True);
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(10, 5))

sns.histplot(cancer_df['mean radius'], kde=True, ax=ax1)
ax1.set_xlabel('Mean Radius')
ax1.set_ylabel('Count')

sns.histplot(cancer_df['mean perimeter'], kde=True, ax=ax2)
ax2.set_xlabel('Mean Perimeter')
ax2.set_ylabel('Count')

sns.histplot(cancer_df['mean area'], kde=True, ax=ax3)
ax3.set_xlabel('Mean Area')
ax3.set_ylabel('Count')

plt.tight_layout()
plt.show()
#we need to check that mean radius, mean perimeter, mean area values for the distribution is normal or not.
#we can observe from the hisplot that distibution is right-skewed.
```
